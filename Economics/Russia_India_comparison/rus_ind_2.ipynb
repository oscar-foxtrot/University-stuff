```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

df = pd.read_excel("WEOOct2023all.xlsx")
df = df[(df["Country"] == "Russia") | (df["Country"] == "India")]
```

```python
rus_gdp = df[(df["Country"] == "Russia") & (df["WEO Subject Code"] == "GGXWDG_NGDP")].loc[:, 1992:2022].iloc[0, :]

ind_gdp = df[(df["Country"] == "India") & (df["WEO Subject Code"] == "GGXWDG_NGDP")].loc[:, 1992:2022].iloc[0, :]

ind_gdp.to_numpy()

ind_gdp.index.to_numpy()

fig, ax = plt.subplots(figsize=(8, 6))
fig.patch.set_facecolor('grey')
ax.set_facecolor('grey')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

ax.plot(rus_gdp.index, rus_gdp, linewidth=3, label="Russia", color="red")
ax.plot(ind_gdp.index, ind_gdp, linewidth=3, label="India", color="orange")
ax.title.set_color('white')

img = plt.imread('rus.png')
imagebox = OffsetImage(img, zoom=0.03)
rusim = AnnotationBbox(imagebox, (0.7 - 0.2, 0.365 + 0.013), xycoords='figure fraction', frameon=False)
ax.add_artist(rusim)

img = plt.imread('ind.png')
imagebox = OffsetImage(img, zoom=0.02)
rusim = AnnotationBbox(imagebox, (0.685, 0.775 + 0.011), xycoords='figure fraction', frameon=False)
ax.add_artist(rusim)

fig.text(0.79 - 0.2, 0.2, 'Russia', color='red', fontsize=18)
fig.text(0.78, 0.81 - 0.2, 'India', color='orange', fontsize=18)

fig.text(0.1, -0.07, f'Russia general government gross debt, 2022: {rus_gdp[2022]} %', color='white', fontsize=18)
fig.text(0.1, -0.14, f'India general government gross debt, 2022: {ind_gdp[2022]} %', color='white', fontsize=18)

plt.title("General government gross debt, in % of GDP")
plt.grid(True, color='dimgrey')

plt.savefig("world_data_gengrossdebt.png", bbox_inches='tight')
plt.show()
```
